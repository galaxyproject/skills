# Average of Column 3 from Dataset #13437

This notebook reads dataset #13437 from Galaxy history and computes the average of column 3.

```python
import gxy
import pandas as pd
```

```python
# Download dataset #13437 from Galaxy history
path = await gxy.get(13437)
```

```python
# Read into pandas dataframe
df = pd.read_csv(path, sep="\t", header=None)
df.head()
```

```python
# Compute average of column 3 (0-indexed as column 2)
avg_col3 = df[2].mean()
print(f"Average of column 3: {avg_col3}")
```
